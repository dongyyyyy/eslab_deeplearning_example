## Hue : 색조 (색의 속성)
## Saturation : 채도 ( 색이 탁하고 흐리거나 선명한 정도 )
## Contrast : 대비 (밝은 부분과 어두운 부분의)
## brightness : 밝기

```python
width = 10
height = 10

column = 4
row = 3

plt.figure(figsize=(width,height))
plt.subplot(row, column, 1)

imgplot = plt.imshow(img)
plt.title('Original Image')
# plt.show()

plt.subplot(row, column, 2)
transform = transforms.RandomVerticalFlip(p=1)
new_img = transform(img)
imgplot = plt.imshow(new_img)

plt.title('Vertical Flip Image')
# plt.show()


plt.subplot(row, column, 3)
transform = transforms.RandomHorizontalFlip(p=1)
new_img = transform(img)
imgplot = plt.imshow(new_img)

plt.title('Horizontal Flip Image')


plt.subplot(row, column, 4)
transform = transforms.GaussianBlur(3, sigma=(0.1, 2.0))
new_img = transform(img)
imgplot = plt.imshow(new_img)

plt.title('GaussianBlur filter = 3 Image')

plt.subplot(row, column, 5)
transform = transforms.GaussianBlur(5, sigma=(0.1, 2.0))
new_img = transform(img)
imgplot = plt.imshow(new_img)

plt.title('GaussianBlur filter = 5 Image')
# plt.show()

plt.subplot(row, column, 6)
transform =transforms.RandomAdjustSharpness(2, p=0.5)
new_img = transform(img)
imgplot = plt.imshow(new_img)

plt.title('Sharpeness factor = 2 Image')


plt.subplot(row, column, 7)
transform =transforms.RandomAdjustSharpness(5, p=0.5)
new_img = transform(img)
imgplot = plt.imshow(new_img)

plt.title('Sharpeness factor = 5 Image')
# plt.show()


plt.subplot(row, column, 8)
transform = transforms.ColorJitter(brightness=1, contrast=0, saturation=0, hue=0)
new_img = transform(img)
imgplot = plt.imshow(new_img)
plt.title('brightness Image')

plt.subplot(row, column, 9)
transform = transforms.ColorJitter(brightness=0, contrast=1, saturation=0, hue=0)
new_img = transform(img)
imgplot = plt.imshow(new_img)
plt.title('contrast Image')

plt.subplot(row, column, 10)
transform = transforms.ColorJitter(brightness=0, contrast=0, saturation=1, hue=0)
new_img = transform(img)
imgplot = plt.imshow(new_img)
plt.title('saturation Image')

plt.subplot(row, column, 11)
transform = transforms.ColorJitter(brightness=0, contrast=0, saturation=0, hue=0.5)
new_img = transform(img)
imgplot = plt.imshow(new_img)
plt.title('hue Image')
plt.show()
```
